```python
import DBconnection
cursor, conn = DBconnection.get_connection()

cursor.execute("""create table agency (
    index bigint,
    agency_id bigint,
    agency_name varchar(255),
    agency_url varchar(255),
    agency_timezone varchar(255),
    agency_lang varchar(255),
    agency_phone varchar(255),
    agency_fare_url varchar(255),
    agency_email varchar(255),
    feed_version bigint)""")
conn.commit() 


cursor.execute("""create table alerts (
    level_0 bigint,
    index bigint,
    id varchar(255),
    cause varchar(255),
    effect bigint,
    severity_level bigint,
    agency_id varchar(255),
    route_id bigint,
    route_type bigint,
    stop_id varchar(255),
    direction_id varchar(255),
    text text,
    language varchar(255),
    description_text text,
    feed_version bigint,
    feed_id bigint)""")
conn.commit() 


cursor.execute("""create table calendar (
    index bigint,
    service_id varchar(255),
    monday varchar(255),
    tuesday varchar(255),
    wednesday varchar(255),
    thursday varchar(255),
    friday varchar(255),
    saturday varchar(255),
    sunday varchar(255),
    start_date varchar(255),
    end_date varchar(255),
    feed_version bigint)""")
conn.commit() 


cursor.execute("""create table feed_info (
    index bigint,
    feed_publisher_name varchar(255),
    feed_publisher_url varchar(255),
    feed_lang varchar(255),
    default_lang varchar(255),
    feed_start_date bigint,
    feed_end_date bigint,
    feed_version bigint,
    feed_contact_email varchar(255),
    feed_info_index bigint)""")
conn.commit()


cursor.execute("""create table routes (
    index bigint,
    route_id bigint,
    agency_id bigint,
    route_short_name varchar(255),
    route_long_name varchar(255),
    route_desc varchar(255),
    route_type bigint,
    route_url varchar(255),
    route_color varchar(255),
    route_text_color varchar(255),
    route_sort_order bigint,
    feed_version bigint)""")
conn.commit()


cursor.execute("""create table shapes (
    index bigint,
    shape_id bigint,
    shape_pt_lat double precision,
    shape_pt_lon double precision,
    shape_pt_sequence bigint,
    shape_dist_traveled double precision,
    feed_version bigint)""")
conn.commit()


cursor.execute("""create table stop_times (
    index bigint,
    trip_id varchar(255),
    arrival_time varchar(255),
    departure_time varchar(255),
    stop_id bigint,
    stop_sequence bigint,
    pickup_type bigint,
    drop_off_type bigint,
    timepoint bigint,
    shape_dist_traveled double precision,
    feed_version bigint)""")
conn.commit()


cursor.execute("""create table stops (
    index bigint,
    stop_id bigint,
    stop_code bigint,
    stop_name varchar(255),
    stop_desc varchar(255),
    stop_lat double precision,
    stop_lon double precision,
    stop_url text,
    location_type bigint,
    wheelchair_boarding bigint,
    platform_code text,
    feed_version bigint)""")
conn.commit()


cursor.execute("""create table trip_updates (
    index bigint,
    trip_id varchar(255),
    start_date text,
    route_id text,
    direction_id bigint,
    stop_sequence bigint,
    departure_delay bigint,
    departure_time bigint,
    arrival_delay bigint,
    arrival_time bigint,
    stop_id text,
    feed_version bigint)
""")
conn.commit()


cursor.execute("""create table trips (
    index bigint,
    route_id bigint,
    service_id varchar(255),
    trip_id varchar(255),
    trip_headsign varchar(255),
    direction_id bigint,
    direction text,
    block_id bigint,
    shape_id bigint,
    wheelchair_accessible bigint,
    branch_letter text,
    boarding_type bigint,
    feed_version bigint)""")
conn.commit()

cursor.execute("""create table vehicle_positions (
    index bigint,
    id varchar(255),
    vehicle_trip_trip_id varchar(255),
    vehicle_trip_start_date varchar(255),
    vehicle_trip_route_id bigint,
    vehicle_trip_direction_id bigint,
    lat double precision,
    lon double precision,
    vehicle_positions_bearing double precision,
    vehicle_timestamp double precision,
    vid varchar(255),
    vehicle_vehicle_label varchar(255),
    vehicle_position_speed double precision,
    feed_version bigint,
    feed_id bigint)""")
conn.commit()


cursor.execute("""create table vehicles (
    index bigint,
    vehicle_id bigint,
    vehicle_label varchar(255),
    vehicle_description varchar(255),
    seated_capacity bigint,
    standing_capacity bigint,
    door_count bigint,
    door_width text,
    low_floor bigint,
    bike_capacity bigint,
    wheelchair_access text,
    num_carriages bigint,
    feed_version bigint)""")
conn.commit()
```
